```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
temps = pd.read_csv('temps.csv')
temps = temps.rename(columns={"1850": "Year", "-0.22": "Temp"})
x=temps['Year']
```

```python
plt.figure(figsize=(7,5),dpi=200)
plt.grid(axis='y')
sns.barplot(x='Year', y='Temp', data=temps)
plt.title('Global Temperatures')
plt.xticks(np.arange(0,180,10))
plt.ylabel('Global Temp in Celcius')
plt.xticks(rotation = 45, ha = 'right');
plt.savefig("temps.png")
```

```python
df = pd.read_csv('hurdat_data.csv')
x = df['Max Wind']

df = df.rename(columns={"Date": "Year"})
df = df.merge(temps)
df = df.drop('Unnamed: 0', axis=1)
df
```

```python
plt.figure(figsize=(7,5), dpi=200)
ax = sns.scatterplot(x='Temp', y='Max Wind', data=df, size='Max Wind',
                sizes=(100,500), alpha=0.8, hue='Max Wind', palette='cool')
ax.set(xlabel='Global Surface Temperatures', ylabel='Hurricane Wind Speeds')
plt.title('Hurricane Wind Speeds vs Global Temperatures 1850-2022')
plt.legend(loc=(1.05,0))
plt.xticks(rotation = 45, ha = 'right');
plt.savefig('wind_speeds.jpg')
```

```python
df = pd.read_csv('hurr_data.csv')
df = df.merge(temps)
df = df.drop('Unnamed: 0', axis=1)
df
```

```python
fig = plt.figure(figsize=(7,5), dpi=200)
sns.scatterplot(x='Temp', y='Highest Saffir- Simpson U.S. Category', data=df,
                size='Highest Saffir- Simpson U.S. Category', sizes=(100,500), alpha=0.8, 
                hue='Highest Saffir- Simpson U.S. Category', palette='cool')
plt.title('Hurricane Categories vs Global Temperatures 1850-2022')
plt.legend(loc=(1.05,0))
plt.xticks(rotation = 45, ha = 'right');
fig.savefig('cat_temp.jpg')
```

```python
df.to_csv('hurr_data.csv')
```
